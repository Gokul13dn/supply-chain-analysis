# DATA HANDLING

```python
# Handle invalid zeros
```

```python
(data == 0).sum() # show how many 0 values in each column
```

```python
dc["TPT_Day_Count"]=dc["TPT_Day_Count"].replace(0,dc["TPT_Day_Count"].mean())
# Alternative - dc.replace({"TPT_Day_Count": {0: dc["TPT_Day_Count"].mean()}}, inplace=True)
```

```python
dc=dc[dc["Weight"]>0]
# alternative - dc["Weight"] = dc["Weight"].replace(0, dc["Weight"].mean())
```

```python
(dc == 0).sum()
```

```python
# Fix data types
```

```python
dc.dtypes
```

```python
dc.loc[:, "Order_ID"]=dc["Order_ID"].astype(int)
```

```python
dc.loc[:,"TPT_Day_Count"]= dc["TPT_Day_Count"].astype(int)
```

```python
dc.dtypes
```

```python
# Remove extra spaces (.str.strip), converts to uppercase (.str.upper)
```

```python
dc.loc[:,"Carrier"] = dc["Carrier"].str.strip().str.upper()
dc.loc[:,"Service_Level"] = dc["Service_Level"].str.strip().str.upper()
dc.loc[:,"Orig_Port"] = dc["Orig_Port"].str.strip().str.upper()
dc.loc[:,"Dest_Port"] = dc["Dest_Port"].str.strip().str.upper()
dc.loc[:,"Plant_Code"] = dc["Plant_Code"].str.strip().str.upper()
```

```python
# Checking negatives
```

```python
dc[dc["Unit_Quant"] < 0]
dc[dc["Weight"] < 0]
dc[dc["TPT_Day_Count"] < 0]
dc[dc["Ship_Ahead_Day_Count"] < 0]
dc[dc["Ship_Late_Day_Count"] < 0]
```

```python
# Remove negative numeric values
```

```python
dc = dc[dc["Unit_Quant"] >= 0]
dc = dc[dc["Weight"] >= 0]
dc = dc[dc["TPT_Day_Count"] >= 0]
dc = dc[dc["Ship_Ahead_Day_Count"] >= 0]
dc = dc[dc["Ship_Late_Day_Count"] >= 0] 
```

```python
dc[dc["Ship_Late_Day_Count"] == 0].shape[0] # (or dc[dc["Ship_Ahead_Day_Count"] == 0].shape[0]) on time or no delay
```

```python
def shipment_status(row):
    if row["Ship_Late_Day_Count"] > 0:
        return "Late"
    elif row["Ship_Ahead_Day_Count"] > 0:
        return "Early"
    else:
        return "On Time"

dc["Shipment_Status"] = dc.apply(shipment_status, axis=1)

dc["Shipment_Status"].value_counts()
```

# VISUALIZATION

```python
plt.figure()
sns.countplot(x="Plant_Code", data=dc)
plt.title("Orders per Plant")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.countplot(x="Carrier", data=dc)
plt.title("Orders by Carrier")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
dc.groupby("Plant_Code")["Unit_Quant"].sum().plot(kind="bar")
plt.title("Quantity by Plant")
plt.xticks(rotation=45)
plt.show()
```
